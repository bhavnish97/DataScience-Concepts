```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_dict = {
    'Rajat': 20,
    'Rohit': 30,
    'Rahul': 5,
    'Karan': 4.5,
    'Himanshu': 12,
    'Kiran': 14
}
```

```python
df = pd.DataFrame(data_dict.items(), columns=['Name', 'Salary'])
```

```python
df
```

```python
plt.style.use('fivethirtyeight')
plt.plot(df['Name'], df['Salary'], color='red')
plt.title("Student's Data")
plt.xlabel('Name')
plt.ylabel('Salary')
```

```python
plt.plot(df['Name'], df['Salary'], linestyle='dotted')
```

```python
plt.plot(df['Name'], df['Salary'], marker='o')
```
