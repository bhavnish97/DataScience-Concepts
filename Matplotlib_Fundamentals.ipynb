```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_dict = {
    'Rajat': 20,
    'Rohit': 30,
    'Rahul': 5,
    'Karan': 4.5,
    'Himanshu': 12,
    'Kiran': 14
}
```

```python
df = pd.DataFrame(data_dict.items(), columns=['Name', 'Salary'])
```

```python
df
```

```python
plt.style.use('fivethirtyeight')
plt.plot(df['Name'], df['Salary'], color='red')
plt.title("Student's Data")
plt.xlabel('Name')
plt.ylabel('Salary')
```

```python
plt.plot(df['Name'], df['Salary'], linestyle='dotted')
```

```python
plt.plot(df['Name'], df['Salary'], marker='o')
```

```python
df = pd.read_csv('batter.csv')
df
```

```python
extract_df = df.head(40)
```

```python
plt.scatter(extract_df['avg'], extract_df['strike_rate'])
plt.xlabel('Avg of Batsman')
plt.ylabel('Strike Rate')
```

```python
# Size based on total runs scored, here size of dot is based on the no of runs sored by batsman
plt.scatter(extract_df['avg'], extract_df['strike_rate'], s=extract_df['runs']/100)
plt.xlabel('Avg of Batsman')
plt.ylabel('Strike Rate')
```

```python
# using plt.plot to do basic scatter plot for faster response
plt.plot(extract_df['avg'], extract_df['strike_rate'],'o')
```
