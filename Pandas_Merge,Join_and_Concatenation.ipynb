```python
import numpy as np
import pandas as pd
```

```python
courses = pd.read_csv('/content/courses.csv')
students = pd.read_csv('/content/students.csv')
nov = pd.read_csv('/content/reg-month1.csv')
dec = pd.read_csv('/content/reg-month2.csv')

matches = pd.read_csv('/content/matches.csv')
delivery = pd.read_csv('/content/deliveries.csv')
```

**Merging/Joins**

```python
registerations = pd.concat([nov,dec], ignore_index=True)
```

```python
registerations
```

```python
students
```

```python
# inner join
registerations.merge(students, on='student_id', how='inner')
```

```python
#left join
registerations.merge(students, on='student_id', how='left')
```

```python
students.merge(registerations, on='student_id', how='left')
```

```python
# right join
students.merge(registerations, on='student_id', how='right')
```

```python
# total revenue by the company
registerations.merge(courses, how='inner', on='course_id')['price'].sum()
```

```python
# find revenue month by month
df = pd.concat([nov,dec], keys=['Nov','Dec']).reset_index()
```

```python
df.merge(courses, on='course_id').groupby('level_0')['price'].sum()
```

```python
# revenue earn in each coures
revenue =registerations.merge(courses, on='course_id').groupby('course_name')['price'].sum().reset_index()
```

```python
revenue.plot(kind='bar')
```
