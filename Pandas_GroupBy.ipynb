```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('/content/imdb-top-1000.csv')
```

```python
movies.head()
```

```python
genres = movies.groupby('Genre')
```

```python
genres.sum()
```

```python
genres.max()
```

```python
# Top 3 genres with highest gross
genres['Gross'].sum().sort_values(ascending=False).head(3)
```

```python
# Top 5 genres with highest avg rating
genres['IMDB_Rating'].mean().sort_values(ascending=False).head(5)
```

```python
# nth row from each group
genres.nth(4)
```

```python
# total no of groups
len(genres)
```

```python
# totlal no. of rows in each columns
genres.size()
```

```python
# to get all the row in a particular group
genres.get_group('Comedy')
```

```python
genres.describe()
```

```python
genres.nunique()
```

```python
# We can specify which aggregation we need to apply on each column
genres.agg({
    'Runtime':'mean',
    'Released_Year':'count',
    'Gross':'sum'
})
```

```python
# if we want to apply particular agg functions for each column
genres.agg(['min', 'max', 'sum'])
```

```python
#loops on groups
for group, data in genres:
  print(group, data)
```

```python
# Highest rated movie for each genre
best_rated = pd.DataFrame(columns=movies.columns)
for genre, data in genres:
  best_rated = best_rated.append(data[data['IMDB_Rating'] == data['IMDB_Rating'].max()])
best_rated
```

```python
# number of movies starting with A for each group
genres.apply(lambda x:x['Series_Title'].str.startswith('A').sum())
```

```python
genres.apply(lambda x:x['IMDB_Rating'].rank(ascending=False))
```
