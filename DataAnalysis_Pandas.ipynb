```python
import numpy as np
import pandas as pd
```

```python
import os
files = [file for file in os.listdir('/content/Data')]

for file in files:
  print(file)
```

```python
all_month_data = pd.DataFrame()

for file in files:
  df = pd.read_csv('/content/Data/'+file)
  all_month_data = pd.concat([all_month_data,df])

all_month_data
```

```python
all_month_data.to_csv("all_month_sales.csv", index=False)
```

```python
all_month_data.isnull().sum()
```

```python
all_month_data.dropna(inplace=True)
```

```python
all_month_data
```

**1. Best month of the sale**

```python
# Best month of the sale
all_month_data['month'] = all_month_data['Order Date'].str[0:2]
```

```python
all_month_data
```

```python
all_month_data['month'].unique()
```

```python
all_month_data = all_month_data[all_month_data['month'] != 'Or']
```

```python
all_month_data['month'].astype('int16')
```

```python
all_month_data['Quantity Ordered'] = all_month_data['Quantity Ordered'].astype('int32')
all_month_data['Price Each'] = all_month_data['Price Each'].astype('float32')
```

```python
all_month_data['total_sale'] = all_month_data['Quantity Ordered']*all_month_data['Price Each']
```

```python
all_month_data
```

```python
all_month_data
```

```python
highest_month_sales = all_month_data.groupby('month').sum().reset_index()[['month','total_sale']].sort_values(by='total_sale',ascending=False)
```

```python
highest_month_sales
```

```python
import matplotlib.pyplot as plt
plt.bar(highest_month_sales['month'],highest_month_sales['total_sale'])
```

**2. Which city has highest sales**

```python
all_month_data['Purchase City'] = all_month_data['Purchase Address'].apply(lambda x:x.split(',')[1])
```

```python
all_month_data['Purchase State'] = all_month_data['Purchase Address'].apply(lambda x:x.split(',')[2].split(' ')[1])
```

```python
all_month_data['Purchase_City_State'] = all_month_data['Purchase City'] + ' ' + all_month_data['Purchase State']
```

```python
all_month_data
```

```python
all_month_data['Purchase_City_State'].nunique()
```

```python
all_month_data['Purchase_City_State'].unique()
```

```python
City_Sales = all_month_data.groupby('Purchase_City_State').sum().reset_index()[['Purchase_City_State', 'total_sale']]
```

```python
City_Sales
```

```python
plt.figure(figsize=(8,8))
plt.bar(City_Sales['Purchase_City_State'], City_Sales['total_sale'])
plt.xticks(rotation=90)
```

**3. What is the best time to show advertisement to maximize the sales**

```python
all_month_data['Order Date'] = pd.to_datetime(all_month_data['Order Date'])
all_month_data['Hour'] = all_month_data['Order Date'].dt.hour
all_month_data['Minutes'] = all_month_data['Order Date'].dt.minute
```

```python
all_month_data
```

```python
Hourly_sales = all_month_data.groupby('Hour').count()
```

```python
Hourly_sales = Hourly_sales.reset_index()
Hourly_sales
```

```python
plt.plot(Hourly_sales['Hour'], Hourly_sales['Minutes'])
plt.xticks(Hourly_sales['Hour'])
plt.grid()
```
