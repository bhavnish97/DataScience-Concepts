# Relational Plots
1. Scatter Plot
2. Line Plot

```python
# Scatter Plots

tips = sns.load_dataset('tips')
```

```python
tips
```

```python
# Axes level function
sns.scatterplot(data=tips, x='total_bill', y = 'tip')
```

```python
# Figure level function
sns.relplot(data=tips, x='total_bill', y = 'tip', kind='scatter', hue='sex', style='time')
```

```python
# Line plot
country = px.data.gapminder()
country
```

```python
India = country[country['country']=='India']
```

```python
India
```

```python
sns.lineplot(data=India, x='year', y='lifeExp')
```

```python
sns.relplot(data=India, x='year', y='lifeExp', kind='line')
```

```python
# facet plot -> work only with figure level function
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', col='sex')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', row='sex')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', col='sex', row='smoker')
```

```python
# Column Wrap -> No of plots in one row
sns.relplot(data=country, x='gdpPercap', y='lifeExp', col='year', col_wrap=4, kind='scatter')
```

# Distribution Plots
1. histplot
2. kdeplot
3. rugplot

```python
# displot -> figure level function
# histplot -> Axes level function
sns.histplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill', kind='hist')
```

```python
# bins
sns.histplot(data=tips, x='total_bill', bins=20)
```

```python
sns.histplot(data=tips, x='day')
```

```python
sns.displot(data=tips, x='total_bill', kind='hist', hue='sex')
```

```python
sns.displot(data=tips, x='total_bill', kind='hist', hue='sex', element='step')
```

```python
# facet plot
sns.displot(data=tips, x='total_bill', kind='hist', col='sex')
```

```python
# kdeplot
sns.displot(data=tips, x='total_bill', kind='kde')
```

```python
sns.displot(data=tips, x='total_bill', kind='kde', hue='sex')
```

```python
# Rugplot -> only have axes level function level
sns.rugplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill', y='tip', kind='kde')
```

```python
sns.displot(data=tips, x='total_bill', y='tip', kind='hist')
```

# Matrix Plot
1. Heatmap
2. Clustermap

```python
country
```

```python
country.pivot(index='country', columns='year', values='lifeExp')
```

```python
# heatmap is axes level function and there is no figure level function for this 
plt.figure(figsize=(12,10))
sns.heatmap(country.pivot(index='country', columns='year', values='lifeExp'))
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(country[country['continent']=='Asia'].pivot(index='country', columns='year', values='lifeExp'), annot=True, lineWidth=0.5)
```

```python
plt.figure(figsize=(12,10))

sns.heatmap(country[country['continent']=='Asia'].pivot(index='country', columns='year', values='lifeExp'), annot=True, lineWidth=0.5, cmap='viridis')
```
